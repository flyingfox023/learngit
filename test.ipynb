```python
import numpy as np
import matplotlib.pyplot as plt

w1 = np.linspace(-1,1,100)
w2 = np.linspace(-1,1,100)

l1_norm = lambda x1,x2: np.abs(x1) + np.abs(x2)
l2_norm = lambda x1,x2:pow(x1,2) + pow(x2,2)
```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

Z = l1_norm(w1,w2)
ax = plt.axes(projection='3d')
ax.plot(w1,w2,Z)
```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

Z = l2_norm(w1,w2)
ax = plt.axes(projection='3d')
ax.plot(w1,w2,Z)
```

```python
%matplotlib inline

X,Y = np.meshgrid(w1,w2)
Z = l1_norm(X,Y)
plt.contourf(X,Y,Z,10)
plt.show()
```

```python
%matplotlib inline

X,Y = np.meshgrid(w1,w2)
Z = l2_norm(X,Y)
plt.contourf(X,Y,Z,10)
plt.show()
```
